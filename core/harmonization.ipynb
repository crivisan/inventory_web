```python
import pandas as pd

# Read the CSV
df = pd.read_csv(path)
df['Komentar_IS_DB'] = None

# Get unique gemeinde values
unique_gemeinden = df['gemeinde'].unique()

# Split and save each gemeinde as separate CSV
for gemeinde in unique_gemeinden:
    # Filter dataframe for current gemeinde
    gemeinde_df = df[df['gemeinde'] == gemeinde]
    
    # Create filename
    filename = f"inventar_{gemeinde}.csv"
    
    # Save to CSV
    gemeinde_df.to_csv(filename, index=False,encoding="utf-8-sig")
    
    print(f"Saved: {filename}")

df.to_csv('00_Inventar_Ganze_Tabelle.csv', index=False, encoding="utf-8-sig")
print(f"Created {len(unique_gemeinden)} files")
```
